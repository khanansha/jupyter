```python
df3=pd.read_csv('C:/Users/akhan.extern/Desktop/jsondata/Hotel_Reviews.csv')
```

```python
uqnhotel=df3.drop_duplicates(subset = ["Hotel_Name"])
```

```python
hotelname=uqnhotel[['Hotel_Address','Hotel_Name','Average_Score','lng','lat']]
hotel_name=hotelname.reset_index(drop=True)
```

```python
df1.columns
```

```python
from geopy.geocoders import Nominatim
```

```python
geolocator = Nominatim(user_agent="geoapiExercises")
def city_state_country(idx,coord):
    location = geolocator.reverse(coord, exactly_one=True)
    address = location.raw['address']
    city = address.get('city', '')
    state = address.get('state', '')
    country = address.get('country', '')
    
   
    hotel_name.loc[idx,'city'] = city
    hotel_name.loc[idx,'state'] = state
    hotel_name.loc[idx,'country'] = country
    return city, state, country

    

for idx, rows in hotel_name.iterrows():
    coord = f"{rows['lat']},{rows['lng']}"
    city_state_country(idx,coord)
```
